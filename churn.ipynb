## Data Preprocessing

```python
spark_df = spark_df.dropna()
```

```python
spark_df = spark_df.withColumn("age_square", spark_df.age**2)
```

```python
spark_df.show(5)
```

```python
from pyspark.ml.feature import StringIndexer 

stringIndexer = StringIndexer(inputCol="churn", outputCol="label")
```

```python
?stringIndexer
```

```python
indexed = stringIndexer.fit(spark_df).transform(spark_df)
```

```python
indexed.dtypes
```

burada churn degiskeni zaten 1 ve 0. ama gercek dunya veri setleri icin bu islemi yapmamiz gerekebilir.

```python
spark_df = indexed.withColumn("label", indexed["label"].cast("integer"))
```

```python
spark_df.dtypes
```

```python
spark_df.groupby("label").count().show()
```

```python
# bagimsiz degiskenlerin ayarlanmasi
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
spark_df.columns
```

```python
bag = ['age', 'total_purchase', 'account_manager', 'years', 'num_sites']
```

```python
vectorAssembler = VectorAssembler(inputCols=bag, outputCol="features")
```

```python
va_df = vectorAssembler.transform(spark_df)
```

```python
va_df.show(3)
```

simdi elimizde hem bagimli bagimsiz degiskenler, hem de onlarin vektorleri var. Simdi final_df olusturalim

```python
final_df = va_df.select(["features", "label"])
```

```python
final_df.show()
```

bazı pyspark algoritmalarının istediği yapılar var. Bu features ve label yapısını kullanarak modellere uygulayabiliyoruz.

## Train-test

```python
splits = final_df.randomSplit([0.7, 0.3])
train_df = splits[0]
test_df = splits[1]
```

```python
train_df.show()
```

```python
test_df.show()
```

## Modelling with GBM

```python
from pyspark.ml.classification import GBTClassifier
```

```python
gbm = GBTClassifier(maxIter=10, featuresCol="features", labelCol="label")
```

```python
gbm_model = gbm.fit(train_df)
```

```python
y_pred = gbm_model.transform(test_df)
```

```python
y_pred.show(4)
```

```python
ac = y_pred.select("label", "prediction")
```

```python
ac.filter(ac.label == ac.prediction).count() / ac.count() # valide edilmemis test hatasi
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

evaluator = BinaryClassificationEvaluator()

paramGrid = (ParamGridBuilder()
    .addGrid(gbm.maxDepth, [2, 4, 6])
    .addGrid(gbm.maxBins, [20, 30])
    .addGrid(gbm.maxIter, [10, 20])
    .build())

cv = CrossValidator(
    estimator=gbm,
    estimatorParamMaps=paramGrid,
    evaluator=evaluator,
    numFolds=10
)
```

```python
cv_model = cv.fit(train_df)
```

```python
y_pred = cv_model.transform(test_df)
```

```python
ac = y_pred.select("label", "prediction")
```

```python
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

```python
evaluator.evaluate(y_pred)
```
